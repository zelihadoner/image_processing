```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
im=plt.imread('manzara.jpg')
plt.imshow(im)
plt.show()
```

```python
im.shape     #üç boyutlu RGB 
```

```python
m,n,p=im.shape
m,n
```

```python
new_im=np.zeros((m,n),dtype=int)
new_im
```

```python
for i in range(m): #resmin satırlarını gezecek
    for j in range(n): #resmin sütunlarını gezecek
        s=(im[i,j,0]+im[i,j,1]+im[i,j,2])/3
        new_im[i,j]=s
        
#bir pikselin rgb değerlerini toplayıp yeni oluşturduğumuz 2 boyutlu matrisin aynı pikseline atıyoruz
###böylece 3 boyutlu resmi iki boyuta indiriyoruz.
```

```python
plt.imshow(new_im,cmap='gray') 
#resim iki boyuta indiği için artık gray level olmuştur bu yüzden resmi cmap=gray olarak bastırıyoruz
plt.show()
plt.imsave('manzara2B.jpg',new_im,cmap='gray')
```

```python
new_im.shape
```

```python
#3 boyutlu resmi 2 boyuta indirgeyip 90 derece döndürme
new_im_90_2b=np.zeros((n,m),dtype=int)
for i in range(m):
    for j in range(n):
            s=(im[i,j,0]+im[i,j,1]+im[i,j,2])/3
            new_im_90_2b[j,i]=s
plt.imshow(new_im_90_2b,cmap='gray')
plt.show()
plt.imsave('manzara2b90.jpg',new_im_90_2b,cmap='gray')
```

```python
#üç boyulu resmi 90 derece döndürme
new_im_90_3b=np.zeros((n,m,p),dtype=np.uint8)
for i in range(m):
    for j in range(n):
        for k in range(p):
            #s=(im[i,j,0]+im[i,j,1]+im[i,j,2])/3   #yanlış çözüm
            new_im_90_3b[j,i,k]=im[i,j,k]
plt.imshow(new_im_90_3b)
plt.show()
plt.imsave('manzara3b90.jpg',new_im_90_3b) 
```
