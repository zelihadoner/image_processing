```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
im_1=plt.imread('test_1.jpg')
im_1.shape
```

```python
im_2=np.zeros((256,512),dtype=int)
m,n=im_2.shape
```

```python
im_2=im_1[:,:,0]
```

```python
plt.imshow(im_2,cmap='gray')
plt.show()
```

```python
im_3=np.zeros((256,512),dtype=int)    #filter uyguladıktan sonra atanacak resim
for i in range(1,m-1):
    for j in range(1,n-1):
        s=(im_2[i-1,j-1]/9+ 
        im_2[i-1,j]/9+
        im_2[i-1,j+1]/9+
        im_2[i,j-1]/9+
        im_2[i,j]/9+
        im_2[i,j+1]/9+
        im_2[i+1,j-1]/9+
        im_2[i+1,j]/9+
        im_2[i+1,j+1]/9)
        s=int(s)
        im_3[i,j]=s
```

```python
plt.subplot(1,2,1)
plt.imshow(im_2,cmap='gray')

plt.subplot(1,2,2)
plt.imshow(im_3,cmap='gray')
```
