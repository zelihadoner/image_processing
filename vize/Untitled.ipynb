```python
import os
import numpy as np
import matplotlib.pyplot as plt
im_ab=plt.imread('ab.png')
plt.imshow(im_ab)
plt.show()
im_cd=plt.imread('cd.png')
plt.imshow(im_cd)
plt.show()
m,n,p=im_ab.shape
m,n,p
im_ab2=np.zeros((m,n),dtype=int)
for i in range(m):
    for j in range(n):
        s=(im_ab[i,j,0]+im_ab[i,j,1]+im_ab[i,j,2])/3
        im_ab2[i,j]=s
im_cd2=np.zeros((m,n),dtype=int)
for i in range(m):
    for j in range(n):
        s=(im_cd[i,j,0]+im_cd[i,j,1]+im_cd[i,j,2])/3
        im_cd2[i,j]=s
im_abcd=np.zeros((150,300),dtype=int)
im_abcd=im_ab2+im_cd2
plt.imshow(im_abcd,cmap='gray')
plt.show()
```
