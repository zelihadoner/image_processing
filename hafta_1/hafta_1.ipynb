```python
m,n,p=im_1.shape
m,n,p
```

```python
plt.imshow(im_1)
plt.show()
```

```python
new_image=np.zeros((m,n),dtype=int)
new_image
```

```python
for i in range(m):
    for j in range(n):
        s=(im_1[i,j,0]+im_1[i,j,1]+im_1[i,j,2])/3
        new_image[i,j]=s
```

```python
plt.imshow(new_image,cmap='gray')
plt.show()
plt.imsave('dubai_gray.jpg',new_image,cmap='gray')
```

```python
new_image2=np.zeros((n,m),dtype=int)
new_image2
```

```python
for i in range(m):
    for j in range(n):
        s=(im_1[i,j,0]+im_1[i,j,1]+im_1[i,j,2])/3
        new_image2[j,i]=s
```

```python
plt.imshow(new_image2,cmap='gray')
plt.show()
plt.imsave('dubai90.jpg',new_image2,cmap='gray')
```
