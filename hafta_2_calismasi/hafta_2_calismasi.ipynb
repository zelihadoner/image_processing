```python
import matplotlib.pyplot as plt
im_1=plt.imread('tokyo.jpg')
plt.imshow(im_1)
plt.show()
```

```python
im_1.ndim,im_1.shape
```

```python
#R bantını 25 eksiltme
m,n,p=im_1.shape   #im_1 i im_2 ye atayabilmek için shape değerlerine ihtiyaç var
im_2=np.zeros((m,n,p),dtype=int)
for i in range(m):
    for j in range(n):
        im_2[i,j,0]=im_1[i,j,0]-125   #R bantını 125 eksilttik 
        im_2[i,j,1]=im_1[i,j,1]    #G ve B bantlarını aynen im_2 ye atadık
        im_2[i,j,2]=im_1[i,j,2]
plt.imshow(im_2)
plt.show()
```

```python
#R bantını s kadar azaltma fonksiyonu
def r_azalt(im_1,s):
    m,n,p=im_1.shape
    im_2=np.zeros((m,n,p),dtype=int)
    for i in range(im_1.shape[0]):     #satırları gezecek
        for j in range(im_1.shape[1]):    #sütunları gezecek
            im_2[i,j,0]=im_1[i,j,0]-s
            im_2[i,j,1]=im_1[i,j,1]
            im_2[i,j,2]=im_1[i,j,2]
    return im_2

im_3=r_azalt(im_1,100)
```

```python
plt.imshow(im_3)
plt.show()
```

```python
#resmin boyutlarını yarı yarıya azaltan fonksiyon
def kucult_yari_2b(im_1):
    m,n,p=im_1.shape   #im_1 in boyut değerlerini atıyoruz
    new_m=int(m/2)   #yarısını aldık ve yenı eni bulduk
    new_n=int(n/2)   #yarısını aldık ve yenı boyu bulduk.küsüratlı çıkmaması için int yaptık
    im_2=np.zeros((new_m,new_n),dtype=int) #yeni resmi yeni boyutlarına göre 2b oluşturduk
    for i in range(new_m):   #yeni resmin satırları dolaşılacak
        for j in range(new_n):   #yeni resmin sütunları dolaşılacak
            s=(im_1[i*2,j*2,0]+im_1[i*2,j*2,1]+im_1[i*2,j*2,2])/3
            im_2[i,j]=int(s)
    return im_2
im_4=kucult_yari_2b(im_1)
```

```python
plt.imshow(im_4,cmap='gray')
plt.show()
```

```python
#benim örneğim:))
def kucult_yari_3b(im_1):
    m,n,p=im_1.shape
    new_m=int(m/2)
    new_n=int(n/2)
    im_2=np.zeros((new_m,new_n,p),dtype=int)
    for i in range(new_m):
        for j in range(new_n):
                im_2[i,j,0]=im_1[i*2,j*2,0]
                im_2[i,j,1]=im_1[i*2,j*2,1]
                im_2[i,j,2]=im_1[i*2,j*2,2]
    return im_2
```

```python
im_7=kucult_yari_3b(im_1)
plt.imshow(im_7)
plt.show()
```

```python
#resimdeki piksel sayısını yarı yarıya indirme
#boyutlarda yarı yarıya inidirilir
#yeni resimedeki bir piksle önceki resimdeki pikselin yarısı yerleştirilir
def olcek_yari(im_1):
    m,n,p=im_1.shape
    new_m=int(m/2)
    new_n=int(n/2)
    im_2=np.zeros((new_m,new_n,p),dtype=int)
    for i in range(new_m):
        for j in range(new_n):
            im_2[i,j,0]=int(im_1[i*2,j*2,0])
            im_2[i,j,1]=int(im_1[i*2,j*2,1])
            im_2[i,j,2]=int(im_1[i*2,j*2,2])
    return im_2
```

```python
im_5=olcek_yari(im_1)
plt.imshow(im_5)
```

```python
image=plt.imread('manzara.jpg')
plt.imshow(image)
plt.show()
```

```python
#3 boyutlu resmi 2 boyuta çevirip renklerini ters çevirme
m,n,p=image.shape
im_2b=np.zeros((m,n),dtype=int)
for i in range(m):
    for j in range(n):
        s=(image[i,j,0]+image[i,j,1]+image[i,j,2])/3
        im_2b[i,j]=s
plt.imshow(im_2b,cmap='gray')
plt.show()
```

```python
im_2b_inv=np.zeros((m,n),dtype=int)
im_2b_inv=255-im_2b
plt.imshow(im_2b_inv,cmap='gray')
plt.show()
```
